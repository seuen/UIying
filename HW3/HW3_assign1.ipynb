```python
# distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
# present each of the 6 distributions

from __future__ import print_function
__author__= 'Siying'
import os
import numpy as np
import pylab as pl
import json
%pylab inline

from scipy.optimize import curve_fit, minimize

# optional for use
# s = json.load( open(os.getenv("PUI2015")+"/lecture/style.json") )
# pl.rcParams.update(s)
```

```python
mn = 100
mysize=2000/(np.array(range(1,100)+[10]))
#mysize=np.random.choice((range(10,2000)),100)
mean = (np.mean(mysize))
std = (np.std(mysize))

print (mysize, mysize.shape)
```

## 1. Chisquare Distribution:

```python
# chisquare(df, size=None)

csq={}
csq['chisq'] = {}

for n in mysize:
    csq['chisq'][n] = np.random.chisquare(mn, size=n)

means = {}
means['chisq']={}
fig_mu_N=pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)

for nn in csq['chisq'].iterkeys():
    means['chisq'][nn]=csq['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn],'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize),max(mysize)],[mn, mn],'k')
```

```python
allmeans=[]
for n in means['chisq']:
    allmeans.append(means['chisq'][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans, bins=20)
pl.xlabel('sample mean of Chisquare distribution sample')
pl.ylabel('N')
```

## 2. Normal (Gaussion) Distribution:

```python
# normal(loc=0.0, scale=1.0, size=None)

gd={}
gd['normal'] = {}

for n in mysize:
    gd['normal'][n] = np.random.normal(loc=mn, size=n)

means = {}
means['normal']={}
fig_mu_N=pl.figure(figsize=(10,10))
axnorm_mu_n = fig_mu_N.add_subplot(111)

for nn in gd['normal'].iterkeys():
    means['normal'][nn]=gd['normal'][nn].mean()
    axnorm_mu_n.plot(nn, means['normal'][nn],'o')
    axnorm_mu_n.set_xlabel('sample size')
    axnorm_mu_n.set_ylabel('sample mean')
    axnorm_mu_n.plot([min(mysize),max(mysize)],[mn, mn],'k')
```

```python
allmeans=[]
for n in means['normal']:
    allmeans.append(means['normal'][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans, bins=20)
pl.xlabel('sample mean of Normal distribution sample')
pl.ylabel('N')
```

## 3. Poisson Distribution

```python
# poisson(lam=1.0, size=None)

ps={}
ps['pois'] = {}

for n in mysize:
    ps['pois'][n] = np.random.poisson(lam=mn,size=n)

means = {}
means['pois']={}
fig_mu_N=pl.figure(figsize=(10,10))
axpoisn_mu_n = fig_mu_N.add_subplot(111)

for nn in ps['pois'].iterkeys():
    means['pois'][nn]=ps['pois'][nn].mean()
    axpoisn_mu_n.plot(nn, means['pois'][nn],'o')
    axpoisn_mu_n.set_xlabel('sample size')
    axpoisn_mu_n.set_ylabel('sample mean')
    axpoisn_mu_n.plot([min(mysize),max(mysize)],[mn, mn],'k')
```
